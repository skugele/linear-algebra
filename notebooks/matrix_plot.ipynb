## 2D Column Space Vector Plot With Applied Transformations

In this section we will demonstrate the following linear transformations

    1. Rotation
    2. Shear
    3. X-axis Stretch
    4. X-axis Shrink
    5. Y-axis Stretch
    6. Y-axis Shrink
    7. Combined XY-axis Stretch
    8. Combined XY-axis Shrink
    9. Rotation + Shear + XY-Stretch
    
Each transformation $T$ will be applied to a matrix $A$. Transformations will have the general form:

$$T = \begin{bmatrix}a & b \\ c & d \end{bmatrix}$$

All examples below use the following matrix $A$:

$$A = \begin{bmatrix} 4 & 0 & 2 & -2 \\ 0 & 5 & 2 & 2 \end{bmatrix}$$ 

```python
A = np.column_stack([[4,0],[0,5],[2,2],[-2,2]])
print(A)
```

### 90 Degree Counter-Clockwise Rotation (Transform Matrix)
$$T = \begin{bmatrix} 0 & 1 \\ -1 & 0 \end{bmatrix}$$


```python
T = np.column_stack([[0,1],[1,0]])
print(T)
```

```python
plot_column_vectors_with_transform_2d(A, T)
```

```python
T = np.column_stack([[1,0.25],[0.25,0.75]])
print(T)
```

```python
plot_column_vectors_with_transform_2d(A, T)
```
